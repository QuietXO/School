```r
# Stiahnite CSV súbor dostupný na 
# https://peter.butka.website.tuke.sk/res/solutions.csv
# a uložte ho do pracovného adresára

file1URL = 'https://peter.butka.website.tuke.sk/res/solutions.csv'
download.file(file1URL, destfile='solution.csv')
```

```r
# Vytvorte dataframe s názvom dfsolutions,
# ktorý bude obsahovať tieto stiahnuté dáta

dfsolutions = read.csv('solution.csv')
```

```r
# Uložte do premennej time hodnoty atribútu time_left z dataframu dfsolutions

time = dfsolutions$time_left
```

```r
# Vypíšte do konzoly iba prvých 10 riadkov dát

head(dfsolutions, 10)
```

```r
# Vypíšte do konzoly informácie o dátach spolu s dátovým typom každého 
# atribútu a náhľadom jeho hodnôt pomocou jedného príkazu

str(dfsolutions)
```

```r
#  Vymažte v dátach všetky záznamy, ktoré obsahujú chýbajúce hodnoty

ori_size = nrow(dfsolutions)
dfsolutions = na.omit(dfsolutions)
new_size = nrow(dfsolutions)
print(paste('Original Size:', ori_size, '| New Size:', new_size))
```

```r
# Transformujte atribút subject_id z dátového typu integer na factor

dfsolutions$subject_id=as.factor(dfsolutions$subject_id)
head(dfsolutions, 1)
```

```r
# Vypíšte do konzoly šiestu hodnotu v poradí z atribútu stop

dfsolutions$stop[6]
```

```r
# Vypíšte do konzoly hodnotu v šiestom riadku a štvrtom stĺpci

dfsolutions[6,4]
```

```r
# Vytvorte dataframe snázvom dfsolutions_numeric, ktorý bude
# obsahovať len numerické atribúty z dataframu dfsolutions

dfsolutions_numeric = dfsolutions[sapply(dfsolutions, is.numeric)]
head(dfsolutions_numeric, 1)
```

```r
# Použite vhodnú funkciu na cyklické spracovanie, ktorá vypočíta priemery
# hodnôt všetkých atribútov uložených v dfsolutions_numeric

attribute_averages = colMeans(dfsolutions_numeric, na.rm = TRUE)
print(attribute_averages)
```

```r
# Vytvorte nový atribút do dfsolutions s názvom time, ktorý bude
# obsahovať hodnotu TRUE, ak je hodnota v stĺpci time_left
# väčšia ako 2200. Inak bude obsahovať hodnotu FALSE.

dfsolutions$time <- dfsolutions$time_left > 2200
dfsolutions$time <- as.logical(dfsolutions$time)
head(dfsolutions[11:15, ])
```

```r
# Vytvorte nový dataframe s názvom dfsolutions_1, ktorý bude obsahovať len dáta,
# kde hodnoty atribútu problem_id sú väčšie ako 200.

dfsolutions_1 = dfsolutions[dfsolutions$problem_id > 200, ]
head(dfsolutions_1)
```

```r
# Usporiadajte dataframe dfsolutions podľa atribútu subject_id
# a uložte ho do dataframu s názvom dfsolutions_2

dfsolutions_2 <- dfsolutions[order(dfsolutions$subject_id), ]
head(dfsolutions_2[18:22, ])
```

```r
# Vypíšte sumarizáciu atribútu time_left (Min, Max, Medián, Priemer, 1Q, 3Q)

summary(dfsolutions$time_left)
```

```r
# Vytvorte boxplot červenej farby, ktorý zobrazí hodnoty numerického
# atribútu time_left rozdelené podľa atribútu answer.

boxplot(time_left ~ answer, data = dfsolutions, col ='red')
```

```r
# Vytvorte (v base grafickom systéme) histogram zelenej farby, ktorý zobrazí
# hodnoty numerického atribútu time_left len záznamov, kde subject_id = 23.
# Do grafu pridajte tiež hrubú, prerušovanú horizontálnu čiaru na hodnote 5.

# Create a histogram of time_left for subject_id = 23
hist(dfsolutions[dfsolutions$subject_id == 23, ]$time_left, col = 'green',
     main = 'Histogram of time_left for subject_id = 23', xlab = 'time_left')

# Add a thick, dashed horizontal line at the value 5
abline(h = 5, col = "red", lty = 2, lwd = 2)
```

```r
# Pomocou systému vykresľovania Lattice vytvorte 5 grafov vedľa seba,
# ktoré budú znázorňovať atribút start v závislosti od atribútu stop.
# Tieto grafy rozdeľte podľa atribútu answer.

# Load the lattice library if it's not already loaded
# install.packages("lattice")
library(lattice)

# Create side-by-side graphs using xyplot
xyplot(start ~ stop | answer, data = dfsolutions,
       layout = c(5, 1),
       main = "Start vs. Stop by Answer",
       xlab = "Stop",
       ylab = "Start",
       type = c("p", "g"),
       auto.key = list(columns = 2))
```

```r
# V rámci systému vykresľovania Ggplot, použite funkciu qplot,
# ktorý bude znázorňovať atribút start v závislosti od atribútu stop,
# pričom grafy rozdeľte podľa time.

# Load the ggplot2 library if it's not already loaded
# install.packages("ggplot2")
library(ggplot2)

# Create a scatter plot using qplot with facets
qplot(x = stop, y = start, data = dfsolutions, color = time,
      main = "Start vs. Stop by Time",
      xlab = "Stop",
      ylab = "Start")
```

```r
# Príklad na funkciu, cyklus a podmienku:
# Vytvorte funkciu s názvom funkcia, ktorá bude obsahovať vstupné premenné c,d,n.

funkcia = function(c, d, n) {
  # Your code here
}

# Vytvorte vektor f, ktorý bude nadobúdať hodnoty od 1 po n s krokom 0.5.

funkcia = function(c, d, n) {
      f = seq(1, n, by = 0.5)
}

# Použitím cyklu prejdite všetky čísla vektora f
# a pomocou vetvenia urobte nasledovne operácie:
# - Ak je aktuálne číslo totožné s c, nahraďte ho na výstupe znakom „A“
# - Ak je aktuálne číslo totožné s d, nahraďte na výstupe znakom „B“
# - Inak vypíšte normálnu hodnotu aktuálneho čísla z vektora f

funkcia = function(c, d, n) {
    f = seq(1, n, by = 0.5)
    
    for (i in 1:length(f)) {
        if (f[i] == c)
            print('A')
        else if (f[i] == d)
            print('B')
        else
            print(f[i])
    }
}

funkcia(2, 4, 5)
```

```r
# Ďalšia úloha na grafy:
# Pre dáta quakes (dáta sú dostupné default v R, alebo ich môžete nájsť
# aj na https://peter.butka.website.tuke.sk/res/quakes.csv)

file2URL = 'https://peter.butka.website.tuke.sk/res/quakes.csv'
download.file(file2URL, destfile='quakes.csv')
dfquakes = read.csv('quakes.csv')
```

```r
# Vytvorte v rámci systémov vykresľovania Base a ggplot nasledovnú
# vizualizáciu - rozdeľte plochu zobrazovania na 3 časti tak,
# aby sa zobrazili 3 grafy pod sebou s nasledovnou náplňou:

# Prvý graf (horný) bude histogram atribútu mag s nastavením 15 hraničných hodnôt,
# pričom bude modrej farby, ako aj s popisom osi x "Magnitude of quakes",
# osi y "Number of quakes", hlavného nadpisu "Histogram of quake magnitudes".
# Pridajte ku grafu vertikálnu prerušovanú čiaru pre 
# hodnotu magnitúdy 4.8 červenej farby a hrúbky 3.

# Create a layout for the 3 plots
layout(matrix(1:3, ncol = 1))

# First Plot (Histogram)
hist(dfquakes$mag, breaks = 15, col = "blue",
     main = "Histogram of quake magnitudes",
     xlab = "Magnitude of quakes",
     ylab = "Number of quakes")

# Add a vertical dashed line for magnitude value 4.8 in red and thickness 3
abline(v = 4.8, col = "red", lty = 2, lwd = 3)

# Second Plot (Boxplot)
boxplot(dfquakes$depth, col = "green",
        main = "Depth of quakes",
        ylab = "Depth")

# Add a horizontal line at depth value 400 in yellow and thickness 5
abline(h = 400, col = "yellow", lwd = 5)

# Third Plot (Dot Plot)
plot(dfquakes$long, dfquakes$lat, pch = 17, col = "red",
     main = "Positions of quakes",
     xlab = "Longitude",
     ylab = "Latitude")

# Add a blue rectangle to enclose the specified region
rect(172, -30, 182, -20, col = "blue", border = "blue", lwd = 3)
```
